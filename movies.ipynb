Comenzamos a desanidar los campos solicitados

```python
colum_name = r"'name': '([^']*)"
df_movies['collection_movies'] = df_movies['belongs_to_collection'].str.extract(colum_name, flags=re.IGNORECASE)
```

```python
df_movies.collection_movies.nunique()
```

```python
df_movies [['belongs_to_collection','collection_movies']]
```

```python
#Creamos una nueva columna para los registros duplicados
df_movies['belong_collection'] = df_movies['collection_movies'].str.split(', ')
#Terminamos de duplicar los registros
df_movies = df_movies.explode('belong_collection').reset_index(drop=True)
#Eliminamos la columna duplicada
df_movies.drop('collection_movies', axis=1, inplace=True)
```

```python
df_movies.head()
```

```python
#Convertimos en lista la columna genres
df_movies ['genres'] = df_movies['genres'].apply(ast.literal_eval)

#Aqui extraemos los nombres de los genres
def extract_genres_name(genres_list):
    genres_name = [genree['name'] for genree in genres_list]
    return', '.join(genres_name)
#aplicamos la funcion que hicimos
df_movies['genree'] = df_movies['genres'].apply(extract_genres_name)

df_movies.head()
```

```python
production_name = r"'name': '([^']*)"
df_movies['production_companie'] = df_movies['production_companies'].str.extract(production_name, flags=re.IGNORECASE)
df_movies.head()
```

```python
language = r"'name': '([^']*)"
df_movies['spoken_language'] = df_movies['spoken_languages'].str.extract(language, flags=re.IGNORECASE)
df_movies.head()
```

```python
countries_name = r"'name': '([^']*)"
df_movies['production_countries'] = df_movies['production_countries'].str.extract(countries_name, flags=re.IGNORECASE)
df_movies['production_countries']
```

```python
#Empezamos a rellenar los valores nulos de las columnas revenue, budget con el nÃºmero 0
df_movies ['revenue'].fillna(0, inplace=True)
df_movies ['budget'].fillna(0, inplace= True)

print('Nulos en revenue', df_movies.revenue.isna().sum())
print('Nulos en budget', df_movies.budget.isna().sum())
```

```python
#Cambiamos el formato de fechas a AAAA-mm-dd
df_movies['release_date'] = pd.to_datetime(df_movies['release_date'], errors='coerce')
df_movies['release_date'] = pd.to_datetime(df_movies['release_date']).dt.strftime('%Y-%m-%d')
df_movies['release_date'].isnull().sum()
```

```python
#Eliminamos los valores nulos del campo release date
df_movies = df_movies.dropna(subset= ['release_date'])
df_movies['release_date'].isnull().sum()







```

```python
#creamos la columna release_year
df_movies['release_date'] = pd.to_datetime(df_movies['release_date'])
df_movies['release_year'] = df_movies['release_date'].dt.year
df_movies[['release_date', 'release_year']].head(10)
```

```python
#Creamos la columna con el retorno de inversiÃ³n (revenue/budget)
df_movies['revenue'] = pd.to_numeric(df_movies['revenue'], errors='coerce')
df_movies['budget'] = pd.to_numeric(df_movies['budget'], errors='coerce')

#Calculamos el retorno de la inversiÃ³n y los datos nulos los rellenamos con el nÃºmero 0
df_movies['return'] = df_movies['revenue'] / df_movies['budget']
df_movies['return'].fillna(0, inplace=True)
df_movies['return']
```

```python
#Verificamos que no quedan datos nulos
df_movies['return'].isnull().sum()
```

```python
def peliculas_idioma(idioma: str):
     # Contamos la cantidad de pelÃ­culas producidas en el idioma especificado
    cantidad = df['original_language'].value_counts()[idioma]
    #Retornamos la cantidad de pelÃ­culas producidas en el idioma especificado.
    return f"{cantidad} peliculas fueron estrenadas en idioma {idioma}"
```

```python
def peliculas_duracion(Pelicula):
    ## filtramos para buscar la pelÃ­cula especÃ­fica y obtener su duraciÃ³n y aÃ±o
    movies_title = df.loc[df['title'] == Pelicula]
    duracion = movies_title['runtime'].values[0]
    year = movies_title['release_year'].values[0]
    
    return f"La pelicula {Pelicula} tiene una duraciÃ³n de {duracion}min y fue estrenada en el AÃ±o: {year}"
```

```python
peliculas_duracion('Jumanji')
```

```python
def franquicia(Franquicia: str):
    # Filtramos para buscar las pelÃ­culas de la franquicia especÃ­fica y obtener la cantidad, la ganancia total y el promedio
    movies_collection = df.loc[df['belong_collection'] == Franquicia]
    cantidad = movies_collection.shape[0]
    ganancia_total = movies_collection['revenue'].sum()
    ganancia_promedio = movies_collection['revenue'].mean()
    
    return f"La franquicia {Franquicia} posee {cantidad} pelÃ­culas, con una ganancia total de {ganancia_total} y una ganancia promedio de {ganancia_promedio}"
```

```python
def peliculas_pais(Pais: str):
    #Filtramos las pelÃ­culas producidas en el paÃ­s especÃ­fico y obtenemos la cantidad de peliculas producidas
    countries = df.loc[df['production_countries'] == Pais]
    cantidad = countries.shape[0]

    return f"Se produjeron {cantidad} pelÃ­culas en el paÃ­s de {Pais}"
```

```python
def get_director(nombre_director):
    # Filtramos las pelÃ­culas dirigidas por el director
    director_names = df[df['creww'].str.contains(nombre_director, case=False, na=False)]
    
    if not director_names.empty:
        # Calculamos el Ã©xito del director a travÃ©s del retorno total
        return_total = director_names['return'].sum()
        
        # Obtenemos informaciÃ³n de cada pelÃ­cula dirigida por el director
        pelicula_dirigida = director_names[['title', 'release_date', 'return', 'budget', 'revenue']]
        
        # Creamos una lista de diccionarios con la informaciÃ³n de cada pelÃ­cula
        peliculas = []
        for _, row in pelicula_dirigida.iterrows():
            pelicula = {
                'title': row['title'],
                'release_date': row['release_date'],
                'return': row['return'],
                'budget': row['budget'],
                'revenue': row['revenue']
            }
            peliculas.append(pelicula)
        
        # Devolvemos la informaciÃ³n en un formato de lista de diccionarios
        return {
            'director': nombre_director,
            'success': return_total,
            'movies': peliculas
        }

    return f'El director {nombre_director} no ha dirigido ninguna pelÃ­cula en este conjunto de datos.'
```

```python
def get_director(nombre_director: str):
    nombre_director = nombre_director.lower()  ##Para ignorar mayusculas y minusculas
    # Relleno los valores NaN en las columnas relevantes con valores adecuados para que no largue error
    df['creww'] = df['creww'].fillna('')
    df['return'] = df['return'].fillna(0)
    # Obtenemos el Ã©xito del director, detalles de las pelÃ­culas, costo y ganancia
    director_names = df[df['creww'].str.contains(nombre_director, case=False)]
    return_total = director_names['return'].sum()
    peliculas = []

    for _, row in director_names.iterrows():
        titulo = row['title']
        fecha_lanzamiento = row['release_date']
        retorno = row['return']
        costo = row['budget']
        ganancia = row['revenue']
        peliculas.append({
            'TÃ­tulo': titulo,
            'Fecha de lanzamiento': fecha_lanzamiento,
            'Retorno Total': retorno,
            'Costo': costo,
            'Ganancia': ganancia
        })

    return {
        'Ã‰xito': return_total,
        'PelÃ­culas': peliculas
    }
```
