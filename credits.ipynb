```python
#Importamos las librerias a usar
import pandas as pd
import numpy as np
import ast
import re
import json
```

```python
#Leemos el archivo csv
df_credits = pd.read_csv('Archivos csv\credits.csv')
df_credits
```

```python
df_credits.isnull().sum()
```

```python
# Convertir la columna 'cast' en un DataFrame utilizando json_normalize()
df_cast = pd.json_normalize(df_credits['cast'])
```

```python
df_credits = pd.concat([df_credits, df_cast], axis=1)
```

```python
df_credits
```

```python
#Convertimos en lista la columna cast
df_credits ['cast'] = df_credits['cast'].apply(ast.literal_eval)

#Aqui extraemos los nombres
def extract_cast_name(cast_list):
    cast_name = [castt['name'] for castt in cast_list]
    return', '.join(cast_name)
#aplicamos la funcion que hicimos
df_credits['castt'] = df_credits['cast'].apply(extract_cast_name)

df_credits.head()
```

```python
#Convertimos en lista la columna crew
df_credits ['crew'] = df_credits['crew'].apply(ast.literal_eval)

#Extraemos los nombres
def extract_crew_name(crew_list):
    crew_name = [creww['name'] for creww in crew_list]
    return', '.join(crew_name)
#aplicamos la funcion
df_credits['creww'] = df_credits['crew'].apply(extract_cast_name)

df_credits.head()
```

```python
df_credits.drop(['cast', 'crew'], axis=1, inplace=True)
```
